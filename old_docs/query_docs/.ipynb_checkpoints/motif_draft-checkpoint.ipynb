```python
import query.egrin2_query as e2q
import pymongo
import weblogolib as wl
import numpy
import corebio
from IPython.display import Image
from IPython.display import SVG

client = pymongo.MongoClient('baligadev')
db = client['eco_db']
```

```python
motif = e2q.find_motifs(db, "2_4_1", "run_bicluster_motif", output_type = "array")
motif
```

```python
data = wl.LogoData.from_counts(corebio.seq.unambiguous_dna_alphabet, motif)
```

```python
options = wl.LogoOptions()
format = wl.LogoFormat(data,options)
format.color_scheme = wl.classic
```

```python
tmp = wl.svg_formatter(data, format)
SVG(tmp)
```
